# Linked List

- Linear data structure
- Stores data in nodes which do not need
to be located in contiguous memory cells
- Each node points to the next
- Last node which usually does not point to anything

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class LinkedList: # Unordered
  def __init__(self):
    self.head = None

  def insert_last(self, data):
    node = Node(data) # Creates a new node

    # Checks if list has anything inside
    if self.head == None: # If no head
      self.head = node # Set new node as head
    else:
      cur = self.head # Set cur to head of list
      while cur.next: # Traverse list to last node
        cur = cur.next
      cur.next = node # Link new node to last node

  def insert_first(self, data):
    node = Node(data) # Creates new node

    if self.head is None: # if empty
      self.head = node # Set new node as head
    else: # not empty
      node.next = self.head # Change pointer to add to front
      self.head = node

  def display(self):
    if self.head: # if not empty
      cur = self.head # cur point to head
      while cur is not None: # while not end of list
        print(cur.data, end=" ") # output data
        cur = cur.next # move to next
      print()
      return
    else: # empty
      return print("Empty")

  def update(self, target, data):
    if self.head is None: # If empty
      print("Empty")
      return
    if self.head: # Not empty
      cur = self.head # cur set to head
      while cur and cur.data != target: # while not end of list & data not target
        cur = cur.next # move to next node
      if cur and cur.data == target: # check if cur is in list and data is target
        cur.data = data # set new data
        print('Updated')
        return
      else: # not in list
        print('Target not found')
        return

  def delete(self, target):
    if self.head is None: # Empty list
      print("Empty")
      return
    if self.head.data == target: # If head is target
      self.head = self.head.next # Set head to next node
      print("Deleted")
      return
    else: # Head not target
      cur = self.head # Set cur as head
      prev = None # Set previous to None
      while cur and cur.data != target: # while in list and data not target
        prev = cur # shift prev to cur node
        cur = cur.next # shift cur to next node
      if cur and cur.data == target: # if data is target
        prev.next = cur.next # Set prev to point after cur / skip cur
        print("Deleted")
        return
      else:
        print("Target not found")
        return

  def search(self, target):
    if self.head is None: # empty
      print("Empty")
      return
    else:
      cur = self.head # Set cur to head
      while cur and cur.data != target: # while data not target
        cur = cur.next # Move to next node
      if cur and cur.data == target: # check if equal
        print("Found")
        return
      else: # cur.data is None (outside list)
        print("Not found")
        return

# main
link_list = LinkedList()

link_list.insert_first(1)
link_list.insert_first(2)
link_list.insert_first(3)
link_list.insert_first(4)
link_list.insert_first(5)

link_list.display()
# -------------------------
# ll2 = LinkedList()

# ll2.insert_last(1)
# ll2.insert_last(2)
# ll2.insert_last(3)
# ll2.insert_last(4)
# ll2.insert_last(5)

# ll2.display()
# -------------------------
link_list.update(2, 3)
link_list.display()
link_list.update(6, 3)
# -------------------------
# link_list.delete(6)
# link_list.display()
# # -------------------------
# link_list.search(6)
```
